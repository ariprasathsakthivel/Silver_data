```python
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

def balance():
    balance_amount=int(input("Enter the Balance of the player\n"))
    return balance_amount

class Deck:
    
    def __init__(self,suits,ranks,values):
        self.suit=suits
        self.rank=ranks
        self.value=values
        self.deck = []
        for suit in self.suit:
            for rank in self.rank:
                d=('{} of {}'.format(rank,suit))
                self.deck.append(d)
        
    def shuffle(self):
        random.shuffle(self.deck)
```

```python
class Hand(Deck):
    
    def __init__(self):
        Deck.__init__(self,suits,ranks,values)
        self.card=[]
        self.value_1=0
    
    def get_deck(self):
        print(self.deck)
    
    def set_deck(self,a):
        self.deck.remove(a)
        
    def get_card(self):
        print(self.card)
        
    def cards(self):
        a=random.choice(self.deck)
        self.set_deck(a)
        self.card.append(a)
        #print(self.card)
    
    def get_value(self):
        for i in self.card:
            for j in self.rank:
                if i.find(j)!=-1:
                    self.value_1=self.value_1+self.value[j]
        return self.value_1
        
class Hit_or_stand(Hand):
    
    def __init__(self):
        Hand.__init__(self)
    
    def hit(self):
        a=random.choice(self.deck)
        super().set_deck(a)
        self.card.append(a)
        print(self.card)
            
    def hit_or_stand(self):
        while True:
            yes=input("Do you want to hit or stand. Choose from below options.\n1.Hit\n2.Stand\n")
            if yes=='Hit':
                self.hit()
            elif yes=='Stand':
                break
            else:
                print("Invalid Input")
```

```python
def betting(bet,bal):
    if bet>0 and bet<=bal:
        bal=bal-bet
    else:
        print("Betting not allowed")
        
def check_win(player_value,dealer_value,bet,bal):
    if player_value>21:
        print('player Bust')
        bal=bal-bet
    elif player_value==21:
        print("You won 1.5 times of your bet amount")
        bal=bal+(bet*1.5)
    elif dealer_value>21:
        print("Dealer Bust")
    elif player_value in range(1,22) and dealer_value in range(1,22):
        if player_value>dealer_value and player_value!=dealer_value:
            print('You won')
            bal=bal+(bet*2)
        elif player_value==dealer_value:
            print("It's a draw")
    else:
        print("You lost")
```

```python
def dealer():
    Deck(suits,ranks,values).shuffle()
    bal=balance()
    bet=int(input("Enter the bet amount\n"))
    betting(bet,bal)
    p1=Hand() and Hit_or_stand()
    d=Hand() and Hit_or_stand()
    for i in range(0,2):
        p1.cards()
        d.cards()
    p1.get_card()
    p1.hit_or_stand()
    if d.get_value()<=16:
        d.hit()
    elif d.get_value()>16:
        d.get_card()
    p1.get_card()
    check_win(p1.get_value(),d.get_value(),bet,bal)
```
