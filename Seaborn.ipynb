```python
sns.jointplot(x='speeding',y='alcohol',data=crash,kind='reg')
```

```python
sns.jointplot(x='speeding',y='alcohol',data=crash,kind='hex')
```

```python
sns.jointplot(x='speeding',y='alcohol',data=crash,kind='scatter')
```

```python
sns.barplot(x='sex',y='tip',data=tips)
```

```python
sns.countplot(x='day',hue='smoker',data=tips)
```

```python
sns.boxplot(x='sex',y='total_bill',hue='smoker',data=tips)
```

```python
sns.violinplot(x='sex',y='tip',hue='day',data=tips)
```
